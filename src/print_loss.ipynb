```python
import os
import numpy as np
file_path = '/Users/alexkim/Desktop/mroberta_cnnzh_epoch_result.txt'
```

```python
f = open(file_path)  
```

```python
lines = f.readlines()   
print(lines[-1])
print(len(lines))
```

```python
vals = []
for i in range(len(lines)):
    if i % 2 != 0:
        continue
    else:
        vals.append(lines[i])
print(vals)
```

```python
loss = []
for val in vals:
    val = val.split(" ")
    loss.append(float(val[7]))
print(loss)
min_loss = np.argmin(loss)
print(min_loss)
```

```python
import matplotlib.pyplot as plt
plt.plot(loss[:50]) # 1
print(len(loss))
min_loss_index = np.argmin(loss[:50])
min_loss = np.min(loss[:50])
print(min_loss_index)
print(lines[37])
print(min_loss)
```
